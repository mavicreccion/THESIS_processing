```python
import pandas as pd
import numpy as np
import datetime
```

```python
dataset = pd.read_csv("weather_ogimet_2015_2016_2017_processed_excel.csv", index_col=['Date'], parse_dates=['Date'], skipinitialspace=True)
dataset
```

# No Precipitation and Weather condition

```python
new_dataset = dataset.drop(['precipitationMM', 'Caption', 'Subcaption'], axis=1)
new_dataset
```

```python
new_dataset = new_dataset.replace('-', np.nan)
new_dataset
```

```python
#convert to float
new_dataset = new_dataset.astype(float)

new_dataset = new_dataset.resample('15T').interpolate(method='linear', axis=0).ffill().bfill()
new_dataset
```

```python
new_dataset.to_csv('interpolated_weather_no_precip.csv')
```

```python
start_date = datetime.datetime.strptime('2015-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
end_date = datetime.datetime.strptime('2016-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
mask = (new_dataset.index >= start_date) & (new_dataset.index <= end_date)
peryear_dataset = new_dataset.loc[mask]
peryear_dataset.to_csv('interpolated_weather_no_precip_2015.csv')
```
