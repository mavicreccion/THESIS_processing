```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("flood_pagasa_2015_2016_2017.csv", index_col=['Time'], parse_dates=['Time'], skipinitialspace=True)
dataset = dataset.drop(dataset.columns[0], axis=1)
dataset = dataset.drop(dataset.columns[0], axis=1)
dataset = dataset.drop(dataset.columns[1], axis=1)
# dataset['Time'] = pd.to_datetime(dataset['Time'])
# dataset = dataset[(dataset['Time'].year >= '2015') & (dataset['Time'] <= '2017')]
dataset
```

```python
dataset = dataset.replace('-', np.nan)

dataset = dataset.reset_index().drop_duplicates(subset='Time', keep='last').set_index('Time')
# dataset = dataset.reindex(pd.date_range("2015-01-01 00:00:00", "2018-01-01 00:00:00"))

# #convert to float
dataset = dataset.astype(float)

dataset = dataset.resample('15T').interpolate(method='linear', axis=0).ffill().bfill()
dataset.tail(200)
```
